```python
import redis
```

```python
conn = redis.StrictRedis(

        host='127.0.0.1',

        port=6379,

        db=1)
```

```python
print ('Set Record:', conn.set("data", "데이터"))
```

```python
print ('Set Record:', conn.set("docs", {"phone number" : "123-123435","name" : "John"}))
```

```python
import json

print ('Set Record:', conn.set("docs", json.dumps({"phone number" : "123-123435","name" : "John"})))
```

```python
print ('Get Record:', str(conn.get("data")))
```

```python
print ('Get Record:', conn.get("data").decode('utf-8'))
```

```python
conn.set("data", "이건 데이터일까?")
print ('Get Record:', conn.get("data").decode('utf-8'))
```

```python
print ('Get Record:', json.loads(conn.get("docs")))
```

```python
import time
def send_sold_email_via_queue(conn, seller, item, price, buyer):
    data = {
      'seller_id': seller,
      'item_id': item,
      'price': price,
      'buyer_id': buyer,
      'time': time.time()
    }
    result = conn.rpush('queue:email', json.dumps(data))
    print(result)
```

```python
import random
import time


seller_list = ["Aaron","Andy","Belita","Charles","Daisy","Duncan","Hubert","Kerri","Lukas","Martin"]
item_list = ["LG gram","V50","LG DIOS","LG ThinQ"," LG tromm"]


while True:
    send_sold_email_via_queue(conn=conn,
                              seller=random.choice(seller_list),
                              item=random.choice(item_list),
                              price=random.randrange(500,1000),
                              buyer="buyer_{}".format(random.randrange(1000000)))
    time.sleep(random.randrange(1,5))
```

```python
def publish_sold_email_via_pubsub(conn, seller, item, price, buyer):
    data = {
      'seller_id': seller,
      'item_id': item,
      'price': price,
      'buyer_id': buyer,
      'time': time.time()
    }
    result = conn.publish('mail_channel',json.dumps(data))
    
    print(result, data)
```

```python
while True:
    publish_sold_email_via_pubsub(conn=conn,
                              seller=random.choice(seller_list),
                              item=random.choice(item_list),
                              price=random.randrange(500,1000),
                              buyer="buyer_{}".format(random.randrange(1000000)))
    time.sleep(random.randrange(1,5))
```
