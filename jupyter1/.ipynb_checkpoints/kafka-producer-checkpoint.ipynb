## 토픽 만들기
#### 카프카가 다른 Message queue와 다른 점은 바로 여러 목적의 데이터들을 topic이라는 개념으로 구분이 가능하다는 것입니다.
#### 그래서 producer를 이용해서 데이터를 카프카에 넣고 consumer를 이용해서 데이터를 카프카에서 가져오기 위해서는 topic이 필요하다.
#### 그러므로 우리는 먼저 토픽을 만드는 방법을 알아야 합니다.
#### 토픽을 만들기 위해서는 replication-factor와 partition 설정이 필요합니다. 

## Replication-factor 
#### broker가 다운될때를 대비해서 복사본을 뒤서 장애를 대응하기 위한 것으로 Replication-factor는 이 복사본의 갯수를 뜻한다.


## partition 
#### 처리해야 할 데이터 양이 늘어서 producer와 consumer가 늘어날때, 토픽이 하나라면 그 많은 데이터양을 감당하지 못할 것이다. 
#### 그래서 그 데이터를 수용할 수 있도록 한 토픽을 여러개로 나누어서 병렬로 처리할 수 있도록 한다.

### adminClient 객체생성
#### 우선 필요한 라이브러리를 로드한 후 adminClient객체를 생성합니다. <br>이떄, 카프카 브로커 bootstrap 서버를 같이 입력합니다. 

```python
from confluent_kafka.admin import AdminClient, NewTopic, NewPartitions
import logging

logging.basicConfig()

a = AdminClient({'bootstrap.servers': 'localhost:9092,localhost:9093,localhost:9094'})
```

### create_topics 함수생성
#### topic들을 만들 수 있는 함수를 만들어줍니다. 
#### NewTopic함수에 replication-factor와 partition 갯수를 입력하여 topic 객체를 만듭니다
#### 토픽 객체는 만들고자 하는 토픽 만큼 리스트를 만듭니다. 
#### adminClient 객체의 create_topics 함수를 이용해서 실제 카프카에 토픽생성을 요청합니다. 
#### 토픽생성 결과를 확인합니다.

```python
def create_topics(a, topics):
    """ Create topics """

    new_topics = [NewTopic(topic, num_partitions=1, replication_factor=2) for topic in topics]
    # Call create_topics to asynchronously create topics, a dict
    # of <topic,future> is returned.
    fs = a.create_topics(new_topics)

    # Wait for operation to finish.
    # Timeouts are preferably controlled by passing request_timeout=15.0
    # to the create_topics() call.
    # All futures will finish at the same time.
    for topic, f in fs.items():
        try:
            f.result()  # The result itself is None
            print("Topic {} created".format(topic))
        except Exception as e:
            print("Failed to create topic {}: {}".format(topic, e))
```

### create_topics 함수 실행
#### topic 이름 리스트를 만든 후, create_topics 함수를 실행시킵니다.

```python
topic_names = ["exampleTopic1","exampleTopic2","exampleTopic3"]

create_topics(a,topics=topic_names)
```

### topic과 broker 정보 리스트 
#### list_topic_metadata 함수를 만들어줍니다.

```python
def list_topic_metadata(a, arg = "all"):
    """ list topics and cluster metadata """

    what = arg

    md = a.list_topics(timeout=10)

    print("Cluster {} metadata (response from broker {}):".format(md.cluster_id, md.orig_broker_name))

    if what in ("all", "brokers"):
        print(" {} brokers:".format(len(md.brokers)))
        for b in iter(md.brokers.values()):
            if b.id == md.controller_id:
                print("  {}  (controller)".format(b))
            else:
                print("  {}".format(b))

    if what not in ("all", "topics"):
        return

    print(" {} topics:".format(len(md.topics)))
    for t in iter(md.topics.values()):
        if t.error is not None:
            errstr = ": {}".format(t.error)
        else:
            errstr = ""

        print("  \"{}\" with {} partition(s){}".format(t, len(t.partitions), errstr))

        for p in iter(t.partitions.values()):
            if p.error is not None:
                errstr = ": {}".format(p.error)
            else:
                errstr = ""

            print("    partition {} leader: {}, replicas: {}, isrs: {}".format(
                p.id, p.leader, p.replicas, p.isrs, errstr))
```

```python
list_topic_metadata(a,"all")
```

## python API adminCilent을 이용한 토픽 삭제
#### 자 이제 만들어놓은 토픽을 adminCilent를 이용해서 삭제해보겠습니다. 

### delete_topics 함수 생성
#### delete역시 adminClient에 있는 delete_topics를 이용합니다. 
#### 이 함수 또한 async로 브로커에 요청을 하기 때문에 실습에서는 결과가 올 때까지 대기하고 결과를 출력한다.

```python
def delete_topics(a, topics):
    """ delete topics """

    # Call delete_topics to asynchronously delete topics, a future is returned.
    # By default this operation on the broker returns immediately while
    # topics are deleted in the background. But here we give it some time (30s)
    # to propagate in the cluster before returning.
    #
    # Returns a dict of <topic,future>.
    fs = a.delete_topics(topics, operation_timeout=30)

    # Wait for operation to finish.
    for topic, f in fs.items():
        try:
            f.result()  # The result itself is None
            print("Topic {} deleted".format(topic))
        except Exception as e:
            print("Failed to delete topic {}: {}".format(topic, e))
```

### delete 함수 실행
#### create와 마찬가지로 삭제할 topic 이름 리스트를 만든 후, delete_topics 함수의 parameter로 입력하여 실행시킵니다.¶

```python
topics_deleted = ["exampleTopic3"]

delete_topics(a,topics_deleted)
```

### 파티션 추가하기
#### 만들어진 토픽의 파티션을 추가해보도록 하겠습니다. 

```python
def create_partitions(a, topics):
    """ create partitions """

    new_parts = [NewPartitions(topic, int(new_total_count)) for
                 topic, new_total_count in zip(topics[0::2], topics[1::2])]

    # Try switching validate_only to True to only validate the operation
    # on the broker but not actually perform it.
    fs = a.create_partitions(new_parts, validate_only=False)

    # Wait for operation to finish.
    for topic, f in fs.items():
        try:
            f.result()  # The result itself is None
            print("Additional partitions created for topic {}".format(topic))
        except Exception as e:
            print("Failed to add partitions to topic {}: {}".format(topic, e))
```

```python
topic_partitions = ["exampleTopic1","2","exampleTopic2",3]
create_partitions(a,topic_partitions)
```

## python API를 이용한 메시지 퍼블리싱

```python
from confluent_kafka import Producer
import sys
```

```python
producer = Producer({'bootstrap.servers': 'localhost:9092,localhost:9093,localhost:9094'})
```

```python
def delivery_report(err, msg):
    """ Called once for each message produced to indicate delivery result.
        Triggered by poll() or flush(). """
    if err is not None:
        print('Message delivery failed: {}'.format(err))
    else:
        print('Message delivered to {} [{}]'.format(msg.topic(), msg.partition()))
```

```python
messages = [
    "this is kafka",
    "Publish and subscribe to streams of records, similar to a message queue or enterprise messaging system.",
    "Store streams of records in a fault-tolerant durable way.",
    "Process streams of records as they occur.",
    "Building real-time streaming data pipelines that reliably get data between systems or applications",
    "Building real-time streaming applications that transform or react to the streams of data",
    "Kafka is run as a cluster on one or more servers that can span multiple datacenters.",
    "The Kafka cluster stores streams of records in categories called topics.",
    "Each record consists of a key, a value, and a timestamp."
]

for i,msg in enumerate(messages):
    try:
        # Produce line (without newline)
        producer.produce("exampleTopic1", msg, callback=delivery_report)
        if i > 3:
            producer.produce("exampleTopic2", msg, callback=delivery_report)

    except BufferError:
        sys.stderr.write('%% Local producer queue is full (%d messages awaiting delivery): try again\n' %
                             len(producer))

    # Serve delivery callback queue.
    # NOTE: Since produce() is an asynchronous API this poll() call
    #       will most likely not serve the delivery callback for the
    #       last produce()d message.
    producer.poll(0)

# Wait until all messages have been delivered
sys.stderr.write('%% Waiting for %d deliveries\n' % len(producer))
producer.flush()
```

### 해당 로그 파일들을 한줄씩 읽어서 각 토픽 카프카로 전달해봅시다 


log file의 위치는 /var/log에 있습니다. 

```python
! ls /var/log
```

```python
! cat /var/log/bootstrap.log
```

```python
topic_names = ["alternatives","bootstrap.log","dpkg.log"]

create_topics(a,topics=topic_names)
```

```python
bootstrap_f = open("/var/log/bootstrap.log","r")
dpkg_f = open("/var/log/dpkg.log","r")
alternatives_f = open("/var/log/alternatives.log","r")
```

```python
while True:
    bootstrap_line = bootstrap_f.readline()
    if bootstrap_line:
        producer.produce("bootstrap", bootstrap_line, callback=delivery_report)
    
    dpkg_line = dpkg_f.readline()
    if dpkg_line:
        producer.produce("dpkg.log", dpkg_line, callback=delivery_report)
        
    alternatives_line = alternatives_f.readline()
    if alternatives_line:
        producer.produce("alternatives.log", alternatives_line, callback=delivery_report)
        
    producer.poll(0)
```
